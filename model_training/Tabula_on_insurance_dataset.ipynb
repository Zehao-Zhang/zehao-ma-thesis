```python
# change tabula to tabula_middle_padding to test middle padding method
from tabula import Tabula
import pandas as pd
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split


original_data = pd.read_csv("Real_Datasets/Insurance_compressed.csv")


# train_data, test_data = train_test_split(original_data, train_size=1138, random_state=42)



# train_data.to_csv("/content/drive/MyDrive/Colab Notebooks/Tabula-main-claude/Real_Datasets/Insurance_train.csv", index=False)
# test_data.to_csv("/content/drive/MyDrive/Colab Notebooks/Tabula-main-claude/Real_Datasets/Insurance_test.csv", index=False)

train_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Tabula-main-claude/Real_Datasets/Insurance_train.csv")
```

```python
### Delete these values by accident, regenerate them
import torch
categorical_columns = ["sex", "children", "sm", "region"]
for sigma_value in [0.01, 0.012, 0.02]:  
    model = Tabula(
        llm='distilgpt2',
        experiment_dir="adult_training",
        batch_size=64,
        epochs=400,
        categorical_columns=categorical_columns,
        use_dp=True,
        clip_coeff=1,
        sigma=sigma_value,
        micro_batch_size=16,
        learning_rate=5e-4
    )


    model.model.load_state_dict(torch.load("pretrained-model/model.pt"), strict=False)
    print(f"Now, it's training under sigma {sigma_value}")
    model.fit(train_data)

    torch.save(model.model.state_dict(), "insurance_training/model_400epoch.pt")

    synthetic_data = model.sample(n_samples=1138)
    sigma_str = f"{sigma_value:.3f}".replace('.', '')
    synthetic_data.to_csv(f"insurance_syn_data/insurance-sigma-{sigma_str}.csv", index=False)
```
